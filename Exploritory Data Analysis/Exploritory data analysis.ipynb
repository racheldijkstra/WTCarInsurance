```python
# make additional dataframe to save columns for visualisation of numeric columns
helpdf = pd.DataFrame()
```

```python
# function for visualation of numeric columns
def numeric_values_plot(new_column_name, column_name, bins_, xlabel, title_numeric_part):
    # Categorize the numeric data into bins
    helpdf[new_column_name] = pd.cut(df[column_name], bins=bins_)

    # Count the occurrences of each category
    category_counts = helpdf[new_column_name].value_counts().sort_index()

    # Create a bar plot based on the categories
    category_counts.plot(kind='bar', color='skyblue')

    # Set labels and title
    plt.xlabel(xlabel)
    plt.ylabel('Amount of accidents')
    plt.title(f'Amount of accidents for each {title_numeric_part} range')

    # Show the plot
    plt.show()   
```

```python
# Define bins for the bar plot and visualise distance column
bins = [0, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350]  
numeric_values_plot('Distance', 'Distance(mi)', bins, 'Distance (miles)', 'distance')
```

```python
# Define bins for the bar plot and visualise distance column
bins = [0, 5, 10, 15, 20, 25, 350]  
numeric_values_plot('Distance_2', 'Distance(mi)', bins, 'Distance (miles)', 'distance')
```

```python
# visualise the amount of accidents around each road/traffic type
road_types = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']
value_counts = df[road_types].sum()

colors = plt.cm.viridis(np.linspace(0, 1, len(value_counts)))

plt.figure(figsize=(14, 6))  # Set the figure size
plt.bar(value_counts.index, value_counts.values, color=colors)
plt.xlabel('Road type')
plt.ylabel('Amount of accidents')
plt.title('Amount of accidents around each road type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.savefig('road_type.png', bbox_inches='tight')  # Save figure
plt.show()
```

Hier kunnen we zien dat er vooral veel ongelukken voorkomen rond traffic signals. Dit zou echter ook zo kunnen zijn omdat deze nou eenmaal veel voorkomen in het verkeer. Ook bij deze features geldt dat er pas conclusies getrokken kunnen worden als we data hebben over hoeveel deze verschillende weg en omgevingstypes in elke staat voorkomen. 

```python
counts = df[['Sunrise_Sunset', 'Civil_Twilight','Nautical_Twilight','Astronomical_Twilight']].apply(lambda x: x.groupby(x).count()).transpose()
```

```python
# Define the positions and width for the bars
positions = np.arange(len(counts.index))
width = 0.25

# Create the bar plot
fig, ax = plt.subplots()
for i, column in enumerate(counts.columns):
    ax.bar(positions + i * width, counts[column], width, label=column)

# Set labels, title, and legend
ax.set_xticks(positions + width * (len(counts.columns) - 1) / 2)
ax.set_xticklabels(counts.index)
ax.set_xlabel('Day/Night')
ax.set_ylabel('Amount of accidents')
ax.set_title('Day/Night vs amount of accidents')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

Er lijken in verhouding toch meer ongelukken te gebeuren op de dag. Dit is niet gek aangezien we eerder al hebben gezien dat de meeste ongelukken gebeuren tijdens de spitsmomenten. Verder zijn er op de dag ook meer mensen aanwezig in het verkeer, wat hoogstwaarschijnlijk de kans op ongelukken vergroot.
